```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel
from scipy.stats import multivariate_normal as mvn, pearsonr
import matplotlib
# font = {"size": 20}
# matplotlib.rc("font", **font)
matplotlib.rcParams["text.usetex"] = False
```

```python
plt.figure(figsize=(5, 5))
for th in np.arange(0.0, 2.0, 0.1):
    theta = th * np.pi
    r = 1
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.arrow(x=0, y=0, dx=x, dy=y, head_width=.05)
plt.show()
```

```python
# Spatial coordinates (uniformly spaced)
x1, x2 = np.meshgrid(np.arange(-5, 5, 0.5), np.arange(-5, 5, 0.5))
X1 = np.vstack([x1.ravel(), x2.ravel()]).T
n = X1.shape[0]

# Gene expression (generate from GP)
p = 100
k = 20

kernel = RBF(length_scale=2) + WhiteKernel(noise_level=1)
covariance = kernel(X1)

## View 1
z_list1 = []
for ii in range(k):
    curr_z = mvn(mean=np.zeros(n), cov=covariance).rvs()
    z_list1.append(curr_z)
    
Z = np.vstack(z_list1).T
W = np.random.normal(size=(k, p))
Y1 = Z @ W

## View 2
z_list2 = []
for ii in range(k):
#     curr_z = mvn(mean=np.zeros(n), cov=covariance).rvs()
    curr_z = z_list1[ii] + np.random.normal(scale=0.1)
    z_list2.append(curr_z)
    
Z = np.vstack(z_list2).T
# W = np.random.normal(size=(k, p))
Y2 = Z @ W
```

```python
plt.scatter(Y1[0, :], Y2[0, :])
```

```python
## Shift view 2 locations slightly
offset = 0.25
X2 = X1 + offset
plt.scatter(X1[:, 0], X1[:, 1], label="View 1")
plt.scatter(X2[:, 0], X2[:, 1], label="View 2")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(13, 5))
plt.subplot(121)
plt.scatter(X2[:, 0], X2[:, 1], c=Y2[:, 0])
plt.colorbar()
plt.xlabel("Spatial coordinate 1")
plt.ylabel("Spatial coordinate 2")
plt.title("Spatial coords colored by gene expression")

plt.subplot(122)
# gpr = GaussianProcessRegressor(kernel=kernel).fit(X2, Y2[:, 0])
gpr = GaussianProcessRegressor().fit(X2, Y2[:, 0])
preds = gpr.predict(X2)
plt.scatter(X2[:, 0], X2[:, 1], c=preds)
plt.colorbar()
plt.show()
```

```python
# Compute dense imputation on View 2
gpr = GaussianProcessRegressor().fit(X2, Y2)
x1_full, x2_full = np.meshgrid(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1))
X2_full = np.vstack([x1_full.ravel(), x2_full.ravel()]).T
Y2_full = gpr.predict(X2_full)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X2_full[:, 0], X2_full[:, 1], c=Y2_full[:, 0])
plt.colorbar()
plt.xlabel("Spatial coordinate 1")
plt.ylabel("Spatial coordinate 2")
plt.title("Predicted spatial gene expression for gene 1")
plt.show()
```

```python
point = np.array([0, 0])

corrs = []
dd_list = np.arange(-5, 5, 0.1)
for dd in dd_list:
    X1_idx = np.where((X1 == point).all(axis=1))[0][0]
    X2_adj_idx = np.argmin(((point + dd - X2_full)**2).sum(1))
    X2_idx = np.argmin(((point - X2_full)**2).sum(1))
        
    nonadj_corr = pearsonr(Y1[X1_idx, :], Y2_full[X2_idx])[0]
    adj_corr = pearsonr(Y1[X1_idx, :], Y2_full[X2_adj_idx])[0]
    corrs.append(adj_corr)
```

```python
plt.plot(dd_list, corrs)
plt.axvline(offset, c='r')
plt.show()
```

```python
## Search across range of r and theta

point = np.array([0, 0])

# rval_list = np.arange(-5, 5, 0.1)
# theta_list = np.arange(0, 2*np.pi, 0.1)
# coords = []
# corrs = []
xlist = np.arange(-3, 3, 0.05)
ylist = np.arange(3, -3, -0.05)
corrs = np.zeros((len(xlist), len(ylist)))

# for ii, rval in enumerate(rval_list):
#     for jj, thetaval in enumerate(theta_list):
for ii, xx in enumerate(xlist):
    for jj, yy in enumerate(ylist):
        
#         dd = np.array([
#             rval * np.cos(thetaval),
#             rval * np.sin(thetaval)
#         ])
#         coords.append(dd)
        dd = np.array([xx, yy])

        X1_idx = np.where((X1 == point).all(axis=1))[0][0]
        X2_adj_idx = np.argmin(((point + dd - X2_full)**2).sum(1))
        X2_idx = np.argmin(((point - X2_full)**2).sum(1))

        nonadj_corr = pearsonr(Y1[X1_idx, :], Y2_full[X2_idx])[0]
        adj_corr = pearsonr(Y1[X1_idx, :], Y2_full[X2_adj_idx])[0]
#         corrs.append(adj_corr)
        corrs[jj, ii] = adj_corr
    
```
